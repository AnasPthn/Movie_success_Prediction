```python
# # Assuming 'imdb_score' is the target variable initially
# target = 'imdb_score'

# # Scatter plots between numerical features and target
# for col in numerical_cols:
#     if col != target:
#         plt.figure(figsize=(6,4))
#         sns.scatterplot(x=data[col], y=data[target])
#         plt.title(f'{col} vs {target}')
#         plt.show()

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'imdb_score' is the target variable initially
target = 'imdb_score'

# List of numerical columns excluding the target variable
numerical_cols = [col for col in data.select_dtypes(include=[float, int]).columns if col != target]

# Number of numerical features
num_features = len(numerical_cols)

# Set the number of rows and columns for the grid
ncols = 4  # Number of columns in the grid
nrows = (num_features // ncols) + (num_features % ncols > 0)  # Number of rows needed

# Create a figure with a grid of subplots
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(20, 15))
fig.tight_layout(pad=5.0)  # Adjust padding between plots

# Flatten the axes array to make indexing easier
axes = axes.flatten()

# Scatter plots between numerical features and target
for i, col in enumerate(numerical_cols):
    ax = axes[i]
    x_data = data[col]
    y_data = data[target]
    
    # Apply log transformation if the data is skewed or large
    if col == 'movie_facebook_likes':  # Example for large values
        x_data = np.log1p(x_data)  # Use log1p to handle zero values
        ax.set_xlabel(f'Log of {col}')
    else:
        ax.set_xlabel(col)
    
    # Handle constant values or narrow range
    if col == 'aspect_ratio':  # Example for constant values
        ax.set_ylim(data[target].min(), data[target].max())
    
    sns.scatterplot(x=x_data, y=y_data, ax=ax)
    ax.set_title(f'{col} vs {target}')

# Hide any remaining empty subplots
for j in range(num_features, len(axes)):
    axes[j].axis('off')

plt.show()
```
